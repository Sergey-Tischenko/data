# Обработка данных

## Лидеры по посещаемости среди регионов

Визуализировать данные по всем регионам дело практически невозможное, поэтому выберем самые посещаемые. Для этого сложим все данные по посещаемости за 4 года и сгруппируем по названиям регионов.

Выберем 10 самых посещаемых и создадим из них отдельную таблицу.

```python
top_visits = pd.pivot_table(total_data_museum, values='sum_visits', index=[
    'name_district']).sort_values('sum_visits', ascending=False).head(10)
top_visits
```

С помощью полученной таблицы, рассортируем регионы на основные и остальные. 

```python
pd.options.mode.chained_assignment = None
def sort_districts(row):
    list_top_visits = top_visits.index.to_list()
    if row['name_district'] in (list_top_visits):
        return row['name_district']
    else:
        return 'остальные регионы'


total_data_museum['name_district_new'] = total_data_museum.apply(sort_districts, axis=1)

total_data_museum
```

Создаем новую сводную таблицу уже с учетом новых вариантов названий.

```python
pd.options.display.float_format = '{:.2f}'.format
data_top = pd.pivot_table(total_data_museum, values=['sum_museums', 'items_museum', 'items_exhibited', 'need_restoration', 'sum_visits', 'exhibitions', 'employees', 'money_amt'], index=[
                              'name_district_new', 'year'],aggfunc='median')
data_top 
```

```python
data_top.dtypes
```

```python
data_top = data_top.sort_values('sum_visits', ascending=False)

data_top = data_top.reset_index()
```

Зададим функцию для постройки графиков

```python
def plt_show(data, values, title, ylabel,):
    plt.figure(figsize=(18, 5))
    ax = sns.barplot(x='name_district_new', y=values, hue='year', data=data)
    ax.set(title=title, xlabel='регион', ylabel=ylabel, )
    plt.legend(title='Год', loc='upper right')
    plt.ylim()
    plt.show()
```

# Подробное изучение показателей

Изучим данные по сотрудникам. Для этого в таблице оставим только три колонки

```python
data_employes = data_top[['name_district_new','year', 'employees']]
```

Так как мы будем изучать изменения год от года, зададим новые столбцы

```python
data_employes = data_employes.set_index(
    ['year', 'name_district_new']).unstack(level=0)
data_employes.columns = data_employes.columns.droplevel()
data_employes
```

Даже тут сложно заметить разницу, поэтому посчитаем проценты.

```python
data_employes = data_employes.pct_change(axis='columns').mul(100).fillna(0)
data_employes
```

Так разница видна гораздо лучше. Для визуализации нам не пригодится столбец с 2018 годом, так как мы будем наблюдать изменения, отталкиваясь от него.

```python
data_employes = data_employes.drop(columns=[2018])
```

## Данные о количестве нанятых или уволенных сотрудников

```python
plt.figure(figsize=(14, 5))
cmap = sns.diverging_palette(10, 150, s=100, as_cmap=True)
ax = sns.heatmap(data_employes, cmap=cmap, annot=True, fmt=".1f")
ax.set(title='Уволенные или нанятые сотрудники в процентах',
       xlabel='год', ylabel='регион')


plt.xlim()
plt.show()
```

Зеленым подсвечивается - нанятые сотрудники, так как я посчитал это положительным изменением. Это значит что идет развитие, требуются новые сотрудники.  Красным - уволенные, так как это негативное изменение. 
Во многих регионах в 2019 году активно нанимали сотрудников. Исключения тут Москва, Санкт-Петербург и Краснодарский край. В 2020 году из-за эпидемии количество сотрудников упало во всех регионах, кроме Московской области и остальные регионы. Там вообще наняли больше всего сотрудников в процентном соотношении за весь рассматриваемый период. сложно предположить с чем может быть это связано. Обьяснить открытием одного музея вряд ли это можно. Или тем, что подобрали уволенных из Москвы. там тоже наняли незначительное количество. В 2021 году в Татарстане уволили самое большое количество сотрудников в процентном соотношении за рассматриваемый период. В других регионах количество менялось незначительно. Можно сказать что количество количество сотрудников значительно уменьшилось в год пандемии

```python
data_employes = data_employes.unstack().to_frame(
).reset_index().rename(columns={0: 'percent'})
```

## Данные о количестве экспонатов, требующих реставрацию.

```python
data_restoration = data_top[['name_district_new', 'year', 'need_restoration']]

data_restoration = data_restoration.set_index(
    ['year', 'name_district_new']).unstack(level=0)
data_restoration.columns = data_restoration.columns.droplevel()
data_restoration = data_restoration.pct_change(
    axis='columns').mul(100).fillna(0)
data_restoration = data_restoration.drop(columns=[2018])
data_restoration
```

```python
plt.figure(figsize=(14,5))
cmap = sns.diverging_palette(150, 10, s=100, as_cmap=True)
ax = sns.heatmap(data_restoration, cmap=cmap,  annot=True, fmt=".1f", vmin=-3, vmax=4, center= 0)
ax.set(title='Изменения количества экспонатов, нуждающихся в реставрации ', xlabel='год', ylabel='регион')


plt.xlim()
plt.show()
```

Здесь нельзя однозначно сказать положительно или отрицательно. Но я считаю, что если изменилоось в меньшую сторону - это значит проведена реставрация и экспонат больше не требует её. И красная - это увеличение количества таких в регионе. Значит что еще часть пришла в негодность. Можно сказать что реставрация ведется крайне медленно. А в 2020 году процесс практически остановился. В 2021 году активно проведена реставрация только в Ярославской области. 

```python
data_restoration = data_restoration.unstack().to_frame().reset_index().rename(columns={0:'percent'})
```
